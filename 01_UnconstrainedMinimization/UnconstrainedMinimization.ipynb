# Energy Minimization

In this example we solve the following minimization problem

*Find* $u^* \in H^1_0(\Omega)$ *such that*

$$ u^* = \operatorname*{argmin}_{u \in H^1_0(\Omega)} J(u). $$

Here the energy functional $J(u)$ has the form

$$ J(u) = \frac{1}{2} \int_\Omega k(u) \nabla u \cdot \nabla u \, dx - \int_\Omega f\,u \, dx,$$

where

$$k(u) = k_1 + k_2 u^2, $$

and $k_{1}>0$ and $k_{2}\geq 0$. As will be demonstrated, the optimality conditions for this problem are nonlinear when $k_{2}>0$ and so in general an iterative "outer" optimization procedure (e.g., Gradient descent or Newton's method) must be applied to determine the minimizer $u^*$.

## Necessary optimality condition (Euler-Lagrange condition)

Let $J_{u}(u)(\tilde{u})$ denote the first variation of $J(u)$ with respect to $u$ in *direction* $\tilde{u}$, i.e.

$$J_{u}(u)(\tilde{u}) := \left. \frac{\mathrm{d}}{\mathrm{d} \varepsilon} J(u + \varepsilon \tilde{u})\right|_{\varepsilon=0}.$$

The first-order necessary condition for $u$ to be a minimizer of $J$ is that $J(u)(\tilde{u})$ is equal to $0$ for all admissable variations/directions $\tilde{u}$:

$$ J_{u}(u)(\tilde{u}) = 0 \quad \forall \tilde{u} \in H_0^1(\Omega).$$

### Weak form:

To obtain the weak form of the above necessary condition, we first expand the term $J(u + \varepsilon \tilde{u})$ as

$$J(u + \varepsilon \tilde{u}) = \frac{1}{2} \int_\Omega [k_1 + k_2(u + \varepsilon\tilde{u})^2] \nabla(u + \varepsilon \tilde{u})\cdot\nabla(u + \varepsilon \tilde{u}) dx - \int_\Omega f\,(u+\varepsilon\tilde{u}) dx.$$

We then differentiate

$$
\frac{\mathrm{d}}{\mathrm{d}\varepsilon}J(u+\varepsilon\tilde{u})
= \frac{1}{2}\int_{\Omega}\left[
2\,k_{2}\tilde{u}(u+\varepsilon\tilde{u})
\nabla(u+\varepsilon\tilde{u})\cdot\nabla
(u+\varepsilon\tilde{u})+
(k_{1}+k_{2}(u+\varepsilon\tilde{u})^{2})
2\nabla\tilde{u}\cdot\nabla(u+\varepsilon\tilde{u})
\right]dx-\int_{\Omega}f\,\tilde{u}\,dx,
$$

and set $\varepsilon$ to $0$

$$
\left. \frac{\mathrm{d}}{\mathrm{d}\varepsilon}J(u+\varepsilon\tilde{u})\right|_{\varepsilon=0}=
\frac{1}{2}\int_{\Omega}\left[
2\,k_{2}\tilde{u}\,u
\nabla u \cdot\nabla
 u+
(k_{1}+k_{2}\,u^{2})
2\nabla\tilde{u}\cdot\nabla u
\right]dx-\int_{\Omega}f\,\tilde{u}\,dx
$$

$$
J(u)(\tilde{u})=\int_{\Omega}\left[
k_{2}\tilde{u}\,u
\nabla u \cdot\nabla
 u+
(k_{1}+k_{2}\,u^{2})
\nabla\tilde{u}\cdot\nabla u
\right]dx-\int_{\Omega}f\,\tilde{u}\,dx.
$$

We then write the weak form of the first-order necessary condition for optimality as

*Find* $u\in H_0^1(\Omega)$ *such that*

$$ \int_\Omega \left[k_2 u\, \tilde{u} \nabla u \cdot \nabla u + (k_1 + k_2u^2)\nabla \tilde{u}\cdot \nabla u\right] dx = \int_\Omega f \tilde{u} dx \quad \forall \tilde{u} \in H_0^1(\Omega).$$

### Strong form:
To obtain the strong form, we invoke Green's first identity and write

$$ \int_\Omega \left[k_2 u \nabla u \cdot \nabla u -  \nabla \cdot [(k_1 + k_2u^2) \nabla u] \right] \tilde{u} dx + \int_{\partial \Omega} [(k_1 + k_2u^2) \nabla u]\cdot n \tilde{u} ds = \int_\Omega f \tilde{u} dx \quad \forall \tilde{u} \in H_0^1(\Omega).$$

Since $\tilde{u}$ is arbitrary in $\Omega$ and $\tilde{u} = 0$ on $\partial \Omega$, the strong form of the non-linear boundary problem reads

$$ - \nabla \cdot [(k_1 + k_2u^2) \nabla u] + k_2 u \nabla u \cdot \nabla u = f \quad {\rm in} \; \Omega; $$
$$ u = 0 \quad {\rm on} \; \partial\Omega.$$

## Infinite-dimensional Newton's Method

The infinite-dimensional Newton's method reads

*Given the estimate* $u_{k}$, *of the minimizer of* $J(u)$ *find* $\hat{u} \in H^1_0(\Omega)$ *such that*

$$ J_{u,u}(u_{k})(\tilde{u}, \hat{u}) = -J_{u}(u_{k})(\tilde{u}) \quad \forall \, \tilde{u} \in H_0^1(\Omega).$$

*Update the solution using the Newton direction* $\hat{u}$
$$ u_{k+1} = u_k + \alpha\hat{u}.$$

Where $\alpha>0$ is an appropriately chosen step-length.

### Hessian

To derive the weak form of the Hessian

$$
J_{u,u}(u)(\tilde{u},\hat{u})=\left. \frac{\mathrm{d}}{\mathrm{d}\varepsilon}J_{u}(u+\varepsilon\hat{u})(\tilde{u})\right|_{\varepsilon=0},
$$

we first expand the term $J_{u}(u +\varepsilon \hat{u})(\tilde{u})$ as

$$J_{u}(u+\varepsilon\hat{u})(\tilde{u}) = \int_\Omega \left[k_2 (u+\varepsilon\hat{u}) \tilde{u} \nabla (u+\varepsilon\hat{u}) \cdot \nabla (u+\varepsilon\hat{u}) + (k_1 + k_2(u+\varepsilon\hat{u})^2)\nabla \tilde{u}\cdot \nabla (u+\varepsilon\hat{u}) \right] dx - \int_\Omega f \tilde{u} dx.$$

Then, after some simplification, we obtain

$$J_{u,u}(u)(\tilde{u}, \hat{u}) = \int_\Omega \left[k_2\hat{u}\tilde{u}\nabla u \cdot \nabla u + 2k_2 u \tilde{u} \nabla \hat{u} \cdot \nabla u + 2k_2 u \hat{u} \nabla \tilde{u} \cdot \nabla u + (k_1 + k_2u^2) \nabla \tilde{u} \cdot \nabla \hat{u} \right] dx. $$

### Weak form of Newton step:

*Given* $u \in H_0^1(\Omega)$, *find* $\hat{u} \in H^1_0(\Omega)$ *such that*

$$
\int_\Omega \left[k_2\hat{u}\tilde{u}\nabla u \cdot \nabla u + 2k_2 u \tilde{u} \nabla \hat{u} \cdot \nabla u + 2k_2 u \hat{u} \nabla \tilde{u} \cdot \nabla u + (k_1 + k_2u^2) \nabla \tilde{u} \cdot \nabla \hat{u} \right] dx =
-
\int_\Omega \left[k_2 u \tilde{u} \nabla u \cdot \nabla u + (k_1 + k_2u^2)\nabla \tilde{u}\cdot \nabla u\right] dx - \int_\Omega f \tilde{u} dx \quad \forall\tilde{u}\in H^1_0(\Omega)
$$


The solution is then updated using the Newton direction $\hat{u}$

$$ u^{\rm new} = u + \alpha \hat{u}.$$

Here $\alpha$ denotes a relaxation parameter (back-tracking/line-search) used to achieve global convergence of the Newton method.

### Strong form of the Newton step

$$ - \nabla \cdot \left[ (k_1 + k_2 u^2) \nabla \hat{u}\right] + 2k_2u\nabla\hat{u}\cdot\nabla u - \nabla\cdot(2k_2 u \hat{u} \nabla u) + k_2 \hat{u} \nabla u \nabla u = \nabla \cdot\left[(k_1 + k_2 u^2)\nabla u \right] - k_2 u \nabla u\cdot \nabla u + f \quad {\rm in} \, \Omega.$$
$$ \hat{u} = 0 \quad {\rm on} \, \partial \Omega. $$

## 2. Define the mesh and finite element spaces

We construct a triangulation (mesh) $\mathcal{T}_h$ of the computational domain $\Omega := [0, 1]^2$ with `nx` elements in the *x*-axis direction and `ny` elements in the *y*-axis direction.

On the mesh $\mathcal{T}_h$, we then define the finite element space $V_h \subset H^1(\Omega)$ consisting of globally continuous piecewise linear functions and we create a function $u \in V_h$.

By denoting by $\left[{\phi_i(x)}\right]_{i=1}^{{\rm dim}(V_h)}$ the finite element basis for the space $V_h$ we have
$$ u(x) = \sum_{i=1}^{{\rm dim}(V_h)} \boldsymbol{u}_i \phi_i(x), $$
where $\boldsymbol{u}\in\mathbb{R}^{{\rm dim}(V_h)}$ contains the coefficients for the finite element expansion of $u(x)$.

Finally we define two special types of functions: the `TestFunction` $\hat{u}$ and the `TrialFunction` $\tilde{u}$. These special types of functions are used by `FEniCS` to generate the finite element vectors and matrices which stem from the first and second variations of the energy functional $J(u)$.

```python
nx = 16
ny = 16
mesh = dl.UnitSquareMesh(nx,ny)
Vh = dl.FunctionSpace(mesh, "CG", 1)

uh = dl.Function(Vh)
u_tilde = dl.TestFunction(Vh)
u_hat = dl.TrialFunction(Vh)

nb.plot(mesh, show_axis="on", mytitle='Finite element mesh')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.show()
print("dim(Vh) = ", Vh.dim())
```

## 3. Define the energy functional

We now define the energy functional
$$ J(u) = \frac{1}{2} \int_\Omega (k_1 + k_2u^2) \nabla u \cdot \nabla u dx - \int_\Omega f\,u dx.$$

The parameters $k_1$, $k_2$ and the forcing term $f$ are defined in FEniCS using the keyword [`Constant`](https://fenicsproject.org/olddocs/dolfin/2017.2.0/python/programmers-reference/functions/constant/Constant.html). To define coefficients that are space dependent one should use the keyword [`Expression`](https://fenicsproject.org/olddocs/dolfin/2017.2.0/python/programmers-reference/functions/expression/Expression.html).

The Dirichlet boundary condition
$$ u = 0 \quad {\rm on} \; \partial\Omega$$
is imposed using the [`DirichletBC`](https://fenicsproject.org/olddocs/dolfin/2017.2.0/python/programmers-reference/fem/bcs/DirichletBC.html) class.

To construct this object we need to provide

- the finite element space `Vh`

- the value `u_0` of the solution at the Dirichlet boundary. `u_0` can either be a `Constant` or an `Expression` object.

- the object `Boundary` that defines on which part of $\partial \Omega$ we want to impose such condition.

```python
f  = dl.Constant(1.)
k1 = dl.Constant(0.05)
k2 = dl.Constant(1.)

J = dl.Constant(.5)*(k1 + k2*uh*uh)*dl.inner(dl.grad(uh), dl.grad(uh))*dl.dx - f*uh*dl.dx

class Boundary(dl.SubDomain):
    def inside(self, x, on_boundary):
        return on_boundary

u_0 = dl.Constant(0.)    
bc  = dl.DirichletBC(Vh, u_0, Boundary())
```

## 4. First variation

The weak form of the first variation reads

$$J_{u}(u)(\tilde{u}) = \int_\Omega \left[k_2 u \tilde{u} \nabla u \cdot \nabla u + (k_1 + k_2u^2)\nabla \tilde{u}\cdot \nabla u\right] dx - \int_\Omega f \tilde{u} dx .$$

We use a **finite difference check** to verify that our derivation is correct.
More specifically, we consider a function
$$ u_0 = x(x-1)y(y-1) \in H^1_0(\Omega) $$
and we verify that for a random direction $\tilde{u} \in H^1_0(\Omega)$ we have
$$ r(\varepsilon) := \left| \frac{J(u_0 + \varepsilon \tilde{u}) - J(u_0)}{\varepsilon} - J_{u}(u_0)(\tilde{u})\right| = \mathcal{O}(\varepsilon).$$

In the figure below we show in a loglog scale the value of the residual $r$ of the finite difference approximation of $J_{u}(u_{0})(\tilde{u})$ as a function of $\varepsilon$. We observe that $r$ decays linearly for a wide range of values of $\varepsilon$, however we notice an increase in the error for extremely small values of $\varepsilon$ due to numerical stability and finite precision arithmetic.

**NOTE:** To compute the first variation we can also use the [automatic differentiation](https://fenicsproject.org/olddocs/dolfin/2017.2.0/python/programmers-reference/fem/formmanipulations/derivative.html#dolfin.fem.formmanipulations.derivative) of variational forms capabilities of FEniCS and write

`grad = dl.derivative(J, uh, u_tilde)`

```python
grad = (k2*uh*u_tilde)*dl.inner(dl.grad(uh), dl.grad(uh))*dl.dx + \
       (k1 + k2*uh*uh)*dl.inner(dl.grad(uh), dl.grad(u_tilde))*dl.dx - f*u_tilde*dl.dx

u0 = dl.interpolate(dl.Expression("x[0]*(x[0]-1)*x[1]*(x[1]-1)", degree=2), Vh)

n_eps = 32
eps = 1e-2*np.power(2., -np.arange(n_eps))
err_grad = np.zeros(n_eps)

uh.assign(u0)
J0 = dl.assemble(J)
grad0 = dl.assemble(grad)

uhat = dl.Function(Vh).vector()
uhat.set_local(np.random.randn(Vh.dim()))
bc.apply(uhat)
dir_grad0 = grad0.inner(uhat)

for i in range(n_eps):
    uh.assign(u0)
    uh.vector().axpy(eps[i], uhat) #uh = uh + eps[i]*dir
    Jplus = dl.assemble(J)
    err_grad[i] = abs( (Jplus - J0)/eps[i] - dir_grad0 )

plt.figure()    
plt.loglog(eps, err_grad, "-ob", label="Error Grad")
plt.loglog(eps, (.5*err_grad[0]/eps[0])*eps, "-.k", label=r"First Order, $\propto\varepsilon$")
plt.title("Finite difference check of the first variation")
plt.xlabel(r"$\varepsilon$")
plt.ylabel(r"$r(\varepsilon)$, finite-difference error")
plt.legend(loc = "upper left")
plt.show()
```

## 5. Second variation

The weak form of the second variation reads

$$J_{u,u}(u)(\tilde{u}, \hat{u}) := \left. \frac{d}{d\varepsilon} J_{u}(u+\varepsilon\hat{u})(\tilde{u})\right|_{\varepsilon=0} = 
\int_\Omega \left[k_2\tilde{u}\hat{u}\nabla u \cdot \nabla u + 2k_2 u \hat{u} \nabla \tilde{u} \cdot \nabla u + 2k_2 u \tilde{u} \nabla \hat{u} \cdot \nabla u + (k_1 + k_2u^2) \nabla \hat{u} \cdot \nabla \tilde{u} \right] dx. $$

As before, we verify that for at $u=u_0$ and a random direction $\hat{u} \in H^1_0(\Omega)$ we have
$$ r := \left\| \frac{J_{u}(u_0 + \varepsilon \hat{u})(\tilde{u}) - J_{u}(u_0)(\tilde{u})}{\varepsilon} - J_{u,u}(u_0)(\tilde{u}, \hat{u})\right\| = \mathcal{O}(\varepsilon).$$

$$ r := \left\| \frac{\boldsymbol{g}(u_{0}+\varepsilon \hat{u}) - \boldsymbol{g}(u_{0})}{\varepsilon} - \boldsymbol{H}(u_{0})\boldsymbol{\hat{u}} \right\|_{2} = \mathcal{O}(\varepsilon)$$

where $\boldsymbol{g},\boldsymbol{\hat{u}}\in\mathbb{R}^{{\rm dim}(V_{h})}$, $\boldsymbol{H}\in\mathbb{R}^{{\rm dim}(V_{h})\times {\rm dim}(V_{h})}$ are such that

$$
\left(\boldsymbol{g}(u)\right)_{i} = J_{u}(u)(\phi_{i}) \quad 1\leq i\leq {\rm dim}(V_{h})
$$

$$
\left(\boldsymbol{H}(u)\right)_{i,j} = J_{u,u}(u)(\phi_{i},\phi_{j}) \quad 1\leq i,j\leq {\rm dim}(V_{h}) 
$$

$$
\hat{u}(x) = \sum_{i=1}^{{\rm dim}(V_{h})}
\left(\boldsymbol{\hat{u}}\right)_{i}\phi_{i}(x)
$$

Furthermore

$$
\left(\boldsymbol{H}(u)\boldsymbol{\hat{u}}\right)_{i}=
J_{u,u}(u)(\phi_{i},\hat{u}) \quad 1\leq i \leq {\rm dim}(V_h)
$$


In the figure below we show in a loglog scale the value of $r$ as a function of $\varepsilon$. As before, we observe that $r$ decays linearly for a wide range of values of $\varepsilon$, however we notice an increase in the error for extremely small values of $\varepsilon$ due to numerical stability and finite precision arithmetic.

**NOTE:** To compute the second variation we can also use automatic differentiation and write

`H = dl.derivative(grad, uh, u_hat)`

```python
H = k2*u_hat*u_tilde*dl.inner(dl.grad(uh), dl.grad(uh))*dl.dx + \
     dl.Constant(2.)*(k2*uh*u_tilde)*dl.inner(dl.grad(u_hat), dl.grad(uh))*dl.dx + \
     dl.Constant(2.)*k2*u_hat*uh*dl.inner(dl.grad(uh), dl.grad(u_tilde))*dl.dx + \
     (k1 + k2*uh*uh)*dl.inner(dl.grad(u_tilde), dl.grad(u_hat))*dl.dx

uh.assign(u0)
H_0 = dl.assemble(H)
H_0uhat = H_0 * uhat
err_H = np.zeros(n_eps)

for i in range(n_eps):
    uh.assign(u0)
    uh.vector().axpy(eps[i], uhat)
    grad_plus = dl.assemble(grad)
    diff_grad = (grad_plus - grad0)
    diff_grad *= 1/eps[i]
    err_H[i] = (diff_grad - H_0uhat).norm("l2")
    
plt.figure()    
plt.loglog(eps, err_H, "-ob", label="Error Hessian")
plt.loglog(eps, (.5*err_H[0]/eps[0])*eps, "-.k", label="First Order")
plt.title("Finite difference check of the second variation")
plt.xlabel(r"$\varepsilon$")
plt.ylabel(r"$r(\varepsilon)$, finite-difference error")
plt.legend(loc = "upper left")
plt.show()
```

## 6. The infinite dimensional Newton Method

The infinite dimensional Newton step reads

*Given an objective minimizer estimate* $u_k \in H_0^1(\Omega)$, *find* $\hat{u} \in H^1_0(\Omega)$ *such that*
$$ J_{u,u}(u_k)(\tilde{u}, \hat{u}) =  - J_{u}(u_k)(\tilde{u}) \quad \forall \tilde{u}\in H^1_0(\Omega)$$
Update the solution $u_{k+1}$ using the Newton direction $\hat{u}$
$$ u_{k+1} = u_k + \alpha \,\hat{u}.$$

Here, for simplicity, we choose $\alpha$ equal to 1. In general, to guarantee global convergence of the Newton method the parameter $\alpha$ should be appropriately chosen (e.g. *back-tracking* or *line search*).

The linear systems to compute the Newton directions are solved using the conjugate gradient (CG) with algebraic multigrid preconditioner with a fixed tolerance. In practice, one should solve the Newton system inexactly by early termination of CG 
iterations via EisenstatWalker (to prevent oversolving) and Steihaug (to avoid negative curvature) criteria.

In the output below, for each iteration we report the number of CG iterations, the value of the energy functional, the norm of the gradient, and the inner product between the gradient and the Newton direction $J_{u}(u_k)(\hat{u})$.

In the example, the stopping criterion is relative norm of the gradient $$\frac{
||\boldsymbol{g}(u_{k})||_{2}}{||\boldsymbol{g}(u_0)||_{2}} \leq \tau,$$
where $\tau$ is the relative tolerance `rtol`. However robust implementation of the stopping criterion should also monitor the quantity $\boldsymbol{g}(u_{k})^{\top}\boldsymbol{\hat{u}}$.

```python
uh.assign(dl.interpolate(dl.Constant(0.), Vh))
rtol = 1e-9
max_iter = 10

J0 = dl.assemble(J)
g0 = dl.assemble(grad)
bc.apply(g0)

tol = g0.norm("l2")*rtol

uhat = dl.Function(Vh).vector()
cg_it = 0
print ("{0:3} {1:3}  {2:15} {3:15} {4:15}".format(
      "It", "cg its", "J", "(g, uhat)", "||g||l2") )

for i in range(max_iter):
    [Hn, gn] = dl.assemble_system(H, grad, bc)
    if gn.norm("l2") < tol:
        print ("\nConverged in", i, "Newton iterations and", cg_it, "total conjugate-gradient iterations.")
        break
    mycgits   = dl.solve(Hn, uhat, -1.*gn, "cg", "petsc_amg")
    cg_it = cg_it + mycgits
    uh.vector().axpy(1., uhat)
    Ji = dl.assemble(J)
    print ("{0:3d} {1:3d} {2:15e} {3:15e} {4:15e}".format(
      i, mycgits, Ji, -gn.inner(uhat), gn.norm("l2")) )    
    plt.figure()
    nb.plot(uh, mytitle=r'Minimizer estimate, Newton iteration {0:1d}'.format(i))    
plt.show()
```

Alternatively, we can use the nonlinear solver that is provided in FEniCS. Here one specify the nonlinear form that one would like to solve `grad == 0`, the function that one is solving for `uh`, the essential (Dirichlet) conditions `bc` that `uh` must satisfy, the Jacobian `J` of `grad` (which is here the Hessian `H` and one specify `solver_parameters` to choose the numerical algorithms that are used to approximately determine a stationary point of the energy objective functional $J(u)$. 

```python
uh.assign(dl.interpolate(dl.Constant(0.), Vh))
parameters={"symmetric": True, "newton_solver": {"relative_tolerance": 1e-9, "report": True, \
                                                 "linear_solver": "cg", "preconditioner": "petsc_amg"}}
dl.solve(grad == 0, uh, bc, J=H, solver_parameters=parameters)
final_g = dl.assemble(grad)
bc.apply(final_g)

print ( "Norm of the gradient at converge", final_g.norm("l2") )
print ("Value of the energy functional at convergence", dl.assemble(J) )

nb.plot(uh, show_axis="on", mytitle=r'$u^*(x,y)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.show()
```
